```python
#1. Read the data and display the fields (and data) of the data using Python Pandas library.
import numpy as np
import pandas as pd
# import numpy as np
df = pd.read_csv('data/housing.csv', delimiter = ',')

display(df.info())
display(df.head())
```

```python
print(df.select_dtypes("object").describe().loc['unique', :])
```

```python
unique_ocean_Proximity=df['ocean_proximity'].unique().tolist()
print(unique_ocean_Proximity)
```

```python
null_columns=df.columns[df.isna().any()]
#print(null_columns)

print('Columns with NaN values are:', null_columns)

for c in null_columns:
    if df[c].dtype!='object':
        value = df[c].mean()
    else:
        #print(df[c].mode()[1])
        value = df[c].mode()[0]
        #print (value)
        #value = value[0]  #0 will be the row(s) name
        #print (value)
    df[c].fillna(value,inplace=True)
display(df.info())
display(df)
```

```python
df = pd.get_dummies(df, columns=['ocean_proximity'],drop_first=True) #Remove first level to get k-1 dummies out of k categorical levels.

display(df)
```

```python
from sklearn.model_selection import train_test_split
Xo = df.drop('median_house_value',axis=1).values
y = df.loc[:, 'median_house_value'].values
X = np.c_[np.ones(len(y)), Xo]
X_Tr, X_Ts, y_Tr, y_Ts = train_test_split(X, y, test_size=0.3, random_state=42) 


best_beta = np.linalg.inv(X_Tr.T @ X_Tr) @ X_Tr.T @ y_Tr

print('The closed form estimates are:', np.round(best_beta,2).tolist()) 
```

```python
x_train_nonNormalized=pd.DataFrame(data=X_Tr,columns=['for intercept','longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income','ocean_proximity_INLAND'
                                                               ,'ocean_proximity_ISLAND','ocean_proximity_NEAR BAY'
                                                              ,'ocean_proximity_NEAR OCEAN'])  # 1st row as the column names

x_train_nonNormalized.to_csv('data_Housing/x_train_nonNormalized.csv', index = False, header=True)



x_test_nonNormalized=pd.DataFrame(data=X_Ts,columns=['for intercept','longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income','ocean_proximity_INLAND'
                                                               ,'ocean_proximity_ISLAND','ocean_proximity_NEAR BAY'
                                                              ,'ocean_proximity_NEAR OCEAN'])  # 1st row as the column names

x_test_nonNormalized.to_csv('data_Housing/x_test_nonNormalized.csv', index = False, header=True)


y_train_nonNormalized=pd.DataFrame(data=y_Tr,columns=['median_house_value'])  # 1st row as the column names

y_train_nonNormalized.to_csv('data_Housing/y_train_nonNormalized.csv', index = False, header=True)


y_test_nonNormalized=pd.DataFrame(data=y_Ts,columns=['median_house_value'])  # 1st row as the column names

y_test_nonNormalized.to_csv('data_Housing/y_test_nonNormalized.csv', index = False, header=True)
```
